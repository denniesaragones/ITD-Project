```python
import pandas as pd
import folium
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')

from folium.plugins import HeatMap
```

```python
cc = pd.read_csv('datasets\\Chicago_Crimes.csv')
```

```python
cc.head(4)
```

```python

def normalizer(df,basta_colum):
    normalized  = (df[basta_colum] - df[basta_colum].min()) / \
    (df[basta_colum].max() - df[basta_colum].min())
    return normalized
```

<h2>ARSON HEATMAP<h2>
<h3>Insight 1: The arson heatmap shows a somewhat dispersed pattern with some moderate concentrations on the South and West Sides, potentially indicating vulnerability in areas with vacant properties or socioeconomic challenges.<h3>

```python
arson_filter = cc[cc['Primary Type'] == 'ARSON']

arson_agregate = arson_filter.groupby(['Latitude','Longitude']).size().reset_index(name='crime_count')


arson_agregate['normalized'] = normalizer(arson_agregate,'crime_count')

head_data = arson_agregate[['Latitude','Longitude','normalized']].values.tolist()

mapa = folium.Map(location=[41.7958,-87.6232], zoom_start=10)

HeatMap(head_data).add_to(mapa)

mapa

 
```
